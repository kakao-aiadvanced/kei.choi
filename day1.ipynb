```python
import promptbench as pb

print("all supported datasets: ")
print(pb.SUPPORTED_DATASETS)

dataset_name = 'gsm8k'
dataset = pb.DatasetLoader.load_dataset(dataset_name)

dataset[:3]
```

```python
print("All supported models: ")
print(pb.SUPPORTED_MODELS)

model = pb.LLMModel(model = 'gpt-3.5-turbo', api_key = 'open-ai-key', max_new_tokens = 150)
```

```python
print("All supported methods: ")
print(pb.SUPPORTED_METHODS)

print("Supported datasets for each method: ")
print(pb.METHOD_SUPPORT_DATASET)

method = pb.PEMethod(method='emotion_prompt', dataset=dataset_name, verbose=True, prompt_id = 1)
```

```python
results = method.test(dataset, model, num_samples=3)
results
```

```python
method = pb.PEMethod(method='CoT',
                        dataset=dataset_name,
                        verbose=True,  # if True, print the detailed prompt and response
                        prompt_id = 1  # for emotion_prompt
                        )

results = method.test(dataset,
                      model,
                      num_samples=20# if don't set the num_samples, method will use all examples in the dataset
                      )
results
```

기획문의 사항 -> 기획자에게 문의
개발문의 사항 -> 개발자에게 문의
애매한 사항 -> 주번 문의

```python
from openai import OpenAI
import os

os.environ["OPENAI_API_KEY"] = "open-ai-key"

client = OpenAI()

prompt = """
너는 문의사항을 세가지 분리 사항으로 분리할꺼야
첫번째로 db 쿼리를 통해 개발자가 데이터를 뽑아줘야하는 경우 개발자가 문의사항을 받을꺼야 출력은 'to 개발'로 해줘
두번째로는 정책 관련된 문의사항에 대해서는 기획자에게 문의사항을 쓸꺼야. 출력은 'to 기획' 이야
세번째로 위 두가지 모두 아닌 경우는 'to 주번' 출력할거야

예를 들어 다음과 같은 content와 label을 줄께 이를 기반으로 대답해줘
1. {content: "무상캐시 데이터 추출 부탁 드립니다.", label: "개발" }
2. {content: "세금 계산서 발행 후 수정 발행이 가능한지 궁금합니다.", label: "기획" }
3. {content: "전체 세금 계산서가 얼마나 발행되었는지 수정이 가능한지 궁금합니다.", label: "주번" }
"""

completion = client.chat.completions.create(
    model='gpt-3.5-turbo-0125',
    max_tokens=1024,
    n=1,
    messages=[
        {"role": "system", "content": prompt},
        {"role": "user", "content": "지난 장애가 발생했을 때 생긴 미소진 금액 알려줘"}
    ]
)

print(completion.choices[0].message.content)

```

```python

open_ai_prompt = """
다음 문의사항을 세 가지 카테고리 중 하나로 분류하세요:
1. 개발자
2. 기획자
3. 주번

문의사항이 데이터 추출 또는 기술 개발과 관련된 경우 1(개발자)로 분류하세요.
문의사항이 정책 질문 또는 기획과 관련된 경우 2(기획자)로 분류하세요.
문의사항이 애매하거나 첫 두 카테고리에 명확히 맞지 않는 경우 3(지원)으로 분류하세요.

문의사항: "새 데이터베이스의 데이터 추출 과정에 대해 도와주실 수 있나요?"

분류 결과를 다음 형식으로 출력하세요:
1. to 개발자
"""

completion = client.chat.completions.create(
    model='gpt-3.5-turbo-0125',
    max_tokens=1024,
    n=1,
    messages=[
        {"role": "system", "content": open_ai_prompt},
        {"role": "user", "content": "지난 장애가 발생했을 때 생긴 미소진 금액 알려줘"}
    ]
)

print(completion.choices[0].message.content)
```
